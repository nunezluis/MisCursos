Consideremos el plano determinado por los puntos $P=(1, 1, 1)$, $Q=(2, 4, 6)$ y $R=(2, 2, 2)$. Para construir la ecuación del plano primero ingresamos los vectores posición de cada punto. 

```python
import sympy
from sympy import *
from sympy.vector import *
R = CoordSys3D('R')
```

Dados los vectores

```python
a =  R.i + R.j + R.k
b =  2*R.i + 4*R.j + 6*R.k
c =  2*R.i + 2*R.j + 2*R.k
```

Podemos verificar que los vectores estén en el mismo plano simplemente calculando el triple producto vectorial: ${\bf a}\cdot({\bf b}\times{\bf c})$ entre ellos. Sabemos que si es nulo es porque los vectores son coplanares. 

```python
a.cross(b).dot(c)
```

Necesitamos ahora calcular  los vectores que van del punto $P$ al punto  $Q$ y del punto $P$ al punto $R$, es decir, los vectores: ${\bf ab}={\bf b}-{\bf a}$ y ${\bf ac}={\bf c}-{\bf a}$

```python
ab=b-a
ac=c-a
```

Un vector normal al plano será sencillamente el vector ${\bf n}={\bf ab}\times {\bf ac}$:

```python
n=ab.cross(ac)
n
```

Podemos escoger cualquiera de los vectores originales, en este caso  al vector ${\bf a}$, para escribir la ecuación del plano: ${\bf n}\cdot {\bf r}={\bf n}\cdot {\bf a}$. 

Donde ${\bf r}$ es:

```python
x,y,z  = symbols('x y z') 
r=  x*R.i + y*R.j + z*R.k
```

```python
Eq(n.dot(r),n.dot(a))
```

Probemos a graficar este plano de la manera siguiente

```python
ec=Eq(n.dot(r),n.dot(a))
sol=solve(ec,z)
```

```python
z=sol[0]
z
```

```python
from sympy.plotting import plot3d
plot3d(z, (x, -2, 2), (y, -2, 2))
```
